```python
consent_onthego = pd.DataFrame(columns = ['galaxy', 'cluster', 'something else'])

for index, line in onthego.iterrows():
    if line['total'] != 0:
        consent_onthego = consent_onthego.append({'galaxy': line['galaxy']/line['total'],
                                  'cluster': line['cluster']/line['total'],
                                  'something else': line['something else']/line['total']},ignore_index = True)


```
